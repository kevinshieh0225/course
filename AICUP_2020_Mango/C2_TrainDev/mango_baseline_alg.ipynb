```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split

from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.svm import LinearSVC

from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.metrics import matthews_corrcoef, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
    
def anova_svm_feature_selection(data_x, label_y, num):
    print(f'這是第{i}種損傷的預測')
    # RandomUnderSampler
    rus = RandomUnderSampler(random_state=0)
    data_, label_ = rus.fit_sample(data_x, label_y)
    
    # train-test split
    x_train,x_test,y_train,y_test= train_test_split(data_,label_,test_size=0.3,random_state=0,stratify=label_)
    
    # ANOVA SVM-C
    # 1) anova filter, take 500 best ranked features
    anova_filter = SelectKBest(f_regression, k=500)
    # 2) svm
    svm = LinearSVC(random_state=42, C=1.0, class_weight='balanced')
    
    clf = Pipeline([('anova', anova_filter),
                    ('scaler', StandardScaler()),
                    ('svc', svm)])
    #anova_svm = make_pipeline(anova_filter, clf)
    clf.fit(x_train, y_train)
    y_test_pred = clf.predict(x_test)

    print('evaluating test data...')
    cm = confusion_matrix(y_test, y_test_pred)
    acc = accuracy_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred)
    p = precision_score(y_test, y_test_pred)
    r = recall_score(y_test, y_test_pred)

    print(cm, acc, f1, p, r)
    print()
    print()
    print()
    return clf
```

```python
predict_model = []
for i in range(5):
    anova_svm = anova_svm_feature_selection(data_pd, label_pd.loc[:,i], i)
    predict_model.append(anova_svm)
```

# Predict Dev file image

```python
dev_path = []
with open('./dev.csv', 'r', encoding='utf8') as f:
    for line in f:
        clean_line = re.search('([0-9]+.jpg)',line)
        curr_img_path = f'./Dev/{clean_line[0]}'
        dev_path.append(curr_img_path)
print(f'dev資料集長度為{len(dev_path)}')
```

```python
transform = torchvision.transforms.Compose([
        torchvision.transforms.Resize((224,224)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
        ])

feature_extraction = []
alexnet = models.alexnet(pretrained=True)
vgg16 = models.vgg16(pretrained=True)
for dev_ in  dev_path:
    img = cv2.imread(dev_, cv2.IMREAD_COLOR)
    img = Image.fromarray(img)
    img = transform(img)
    img = img.unsqueeze(0)
    
    with torch.no_grad():
        alex_output = alexnet(img).tolist()
    with torch.no_grad():
        vgg_output = vgg16(img).tolist()
    feature_extraction.append(alex_output[0]+vgg_output[0])
    
print(np.array(feature_extraction).shape)
```

```python
dev_predict = []

for anova_svm_ in predict_model:
    dev_y = anova_svm_.predict(feature_extraction)
    dev_predict.append(dev_y)
```
